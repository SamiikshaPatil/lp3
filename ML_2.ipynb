```python
x=3
lr=0.01
precision=0.000001
previous_state_size=1
max_iters=1000
iters=0
gf=lambda x: (x+3)** 2

```

```python
while precision < previous_state_size and iters<max_iters:
  prev= x
  x=x-lr*gf(prev)
  previous_state_size=abs(x-prev)

  iters=iters+1
  print("\nPrevious state size ",previous_state_size)
  # print("\nIterration",iters,"\nX value is",cur_x)
  print("The local minimum occurs at",x)
```
